Se obtienen las variables numéricas y categóricas

```python
df = pd.read_csv("arboles/housing_train.csv")
df.shape
print("Variables numéricas:")
nums = df.select_dtypes(include=["int64", "float64"]) 
for col in nums.columns:
    print("Los datos estadísticos de la columna " + col + " son: ")
    print(df[col].describe())

print("Variables categóricas:")
cats = df.select_dtypes(exclude=["int64", "float64"])
for col in cats.columns:
        print("La frequencia de la columna " + col + " es de: \n" + str(df[col].value_counts()))

df["SalePrice"].describe()
df["SalePrice"].describe().hist()
```

Se realiza el tratamiento de missing

```python
print("Con nulos")
dataNull=df.isnull().sum()
print([dataNull[dataNull > 0]])

for col in df.columns:
    if (df[col].dtypes == np.int64 or df[col].dtypes == np.float64):
        df[col].fillna(df[col].mean(), inplace=True)
    else:
        df[col].fillna("", inplace=True)

print("Sin nulos")
dataNull=df.isnull().sum()
print([dataNull[dataNull > 0]])

print("Matriz de correlación entre las variables numéricas")
print(nums.corr(method="pearson"))

print("Regresiones:")

print("Regresiones Árbol de decisión:")


print("Regresiones random forest:")


print("Clasificadores:")

print("Clasificación Árbol de decisión:")


print("Clasificación random forest:")
```
